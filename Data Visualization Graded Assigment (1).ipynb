```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the dataset
df = sns.load_dataset('tips')
df.head() 
```

#### Ques1-Create a Line Plot with x (‘total_bill’) versus y (‘tip’) using lineplot() function:

```python
# Create a line plot
sns.lineplot(x='total_bill', y='tip', data=df)

# Set plot labels and title
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Line Plot of Total Bill vs. Tip')

# Show the plot
plt.show()
```

#### Ques2- Create a scatter plot with total bill and tip:

```python
# create a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=df)

# set labels and title
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot of Total Bill vs. Tip')

# show plot
plt.show()
```

#### Ques3- create a hist plot on total_bill with bins=10.
* Create the following -Title as "Frequency of Bill Amount" -xlabel as "Billed Amount" -ylabel as "Count"

```python
# Create a histogram plot
sns.histplot(data=df, x='total_bill', bins=10, kde=False)

# Set plot title, xlabel, and ylabel
plt.title('Frequency of Bill Amount')
plt.xlabel('Billed Amount')
plt.ylabel('Count')

# Show the plot
plt.show()
```

#### Ques4- create a heatmap on tips dataset by showing the corelation between numeric attributes.

```python
# Calculate the correlation matrix for numeric attributes
correlation_matrix = df.corr(numeric_only=True)  # Explicitly set numeric_only to True

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, cmap= 'coolwarm',annot=True,linewidths=1)

# Set plot title
plt.title('Correlation Heatmap of Numeric Attributes')

# Show the plot
plt.show()
```

#### Ques5- create a barplot on smoker and total_bill attribute.

```python
# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='smoker', y='total_bill', data=df)

# Set plot labels and title
plt.xlabel('Smoker')
plt.ylabel('Total Bill')
plt.title('Bar Plot of Total Bill by Smoker')

# Show the plot
plt.show()
```

#### Ques6- use the countplot() function to show the count of smokers and non-smokers:

```python
# Create a countplot for smokers and non-smokers
plt.figure(figsize=(6, 4))
sns.countplot(x='smoker', data=df)

# Set plot labels and title
plt.xlabel('Smoker')
plt.ylabel('Count')
plt.title('Count of Smokers and Non-Smokers')

# Show the plot
plt.show()
```

#### Ques7- analyze how the tipped amount varies during lunch and dinner time and if there are any outliers in our data. Hint :(Boxplot)

```python
# Create a boxplot to compare tipped amounts during lunch and dinner
plt.figure(figsize=(8, 6))
sns.boxplot(x='time', y='tip', data=df)

# Set plot labels and title
plt.xlabel('Time (Lunch vs. Dinner)')
plt.ylabel('Tip Amount')
plt.title('Tipped Amount Variation During Lunch and Dinner')

# Show the plot
plt.show()
```

#### Ques8- create a pair plot to show the pair wise distribution of all attributes

```python
# Create a pair plot
sns.pairplot(df)

# Show the plot
plt.show()
```
