```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
expenditure = pd.read_excel(r"C:\Users\hp\Downloads\6 Practice Exercises\Practice_Exercise_5\CC_expenditure.xlsx")
```

```python
expenditure.head()
```

```python
expenditure["Month"] = expenditure["Date"].dt.month_name()
expenditure["Year"] = expenditure["Date"].dt.year
```

```python
expenditure.head()
```

```python
expenditure.pivot_table(index=["Year","Segment"], values=["Amount spent","Amount repaid"], aggfunc="mean")
```

```python
expenditure["Month_order"]=expenditure["Date"].dt.month
expenditure.head()
```

```python
repay = expenditure.pivot_table(index=["Year","Month_order","Month"],values=["Amount repaid"],aggfunc="mean")
repay
```

```python
sns.set()
repay.plot(figsize=(15,8))
```

```python
total_purchase = expenditure.pivot_table(index=["Type"],columns=["Year"],values=["Amount spent"],aggfunc="sum")
total_purchase
```

```python
total_purchase["Amount spent"].sort_values(by=2004,ascending=False)
```

```python
year2006 = expenditure[expenditure["Year"]==2006]
cc_type = year2006.pivot_table(index=["Month_order","Month"],columns=["Product"],values=["Amount spent"],aggfunc="mean")

sns.set()
cc_type.plot(figsize=(15,8))
```
