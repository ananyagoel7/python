### Multiple Line Plots

```python
plt.figure(figsize=(20,5))
np.random.seed(0) 

summer = np.random.uniform(30,45, size=30)
plt.plot(summer)

winter = np.random.uniform(10,30, size=30)
plt.plot(winter)

rainy = np.random.uniform(20,35, size=30)
plt.plot(rainy)
plt.show()
```

### Adding more properties

```python
# line width linestyle color figsize label loc of legend
plt.figure(figsize=(15,5))
np.random.seed(0) 

summer = np.random.uniform(30,45, size=30)
plt.plot(summer,linestyle="dotted",color="red",lw=3)

winter = np.random.uniform(10,30, size=30)
plt.plot(winter,linestyle="dashed",color="blue",lw=4)

rainy = np.random.uniform(20,35, size=30)
plt.plot(rainy,linestyle="dashdot",color="green",lw=5)

plt.title("Temperature comparison across summer, winter and rainy")
plt.xlabel("Day of the month")
plt.ylabel("Temperature")
plt.legend(["summer","winter","rainy"],loc=1)
plt.show()
```

## Pie chart
* to show percentage or proportion of diff categories 

```python
plt.figure(figsize=(6,6))
values = [140,141,33,14,4]
countries = ["India","China","USA","Russia","Canada"]
plt.pie(values,labels=countries,shadow=True,autopct= '%1.0f%%')
plt.title("Population of different countries")
plt.legend(countries,loc=2)
plt.show()
```

## Scatter plot
* graphically look at the releationship between numerical variables.
* correleation and see clusters if present.

```python
np.random.seed(0)
weight_m = np.random.uniform(60,100, size=30)
height_m = np.random.uniform(150,185,size = 30)
weight_f = np.random.uniform(50,80, size=30)
height_f = np.random.uniform(140,175,size = 30)
plt.scatter(weight_m,height_m,color="red",marker="D")
plt.scatter(weight_f,height_f,color="green",marker="*")
plt.xlabel("Weight in kg",fontweight="bold")
plt.ylabel("Height in cm")
plt.title("Height vs Weight",fontsize=20)
plt.legend(["Male","Female"])
plt.show()
```

## Bar plot
* we can make out small differences easily by comparing the heights of the bars
* we can display a lot of categories in a bar more easily

```python
population = [140,141,33,14,4]
countries = ["India","China","USA","Russia","Canada"]
plt.bar(countries,population,color="lightgreen",edgecolor="black")
plt.title("Population of different countries")
for i in range(len(population)):
    plt.text(i,population[i],population[i],ha="center")
plt.show()
```

### Box and Whisker plot
* used to check for outliers 
* understand the distribution of the dataset by loking the gaps between the outliers

```python
# salaries of different profiles
fresher = np.random.uniform(5,8,size=100)
lead = np.random.uniform(7,12,size=30)
manager = np.random.uniform(8,18,size=20) 
director = np.random.uniform(16,24,size=10)
cxo = np.random.uniform(22,50,size=5)
```

```python
plt.figure(figsize=(15,7))
plt.boxplot([fresher,lead,manager,director,cxo],
            labels=["fresher","lead","manager","director","cxo"],patch_artist=True,vert=0)
plt.show()
```

### Histogram
* useful to understand the distribution of numerical features

```python
values = np.random.uniform(25,100,size=100)
plt.hist(values,bins=15,histtype="bar",color="green",edgecolor="black")
plt.show()
```

## Seaborn

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

### Scatter plot

```python
# theme of the graph
plt.style.use('seaborn-bright')
sns.set_style('whitegrid')
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.xlabel("Total Bill")
plt.show()
```

from the above plot we can see that there is a positive correleation between total bill and tip 

```python
plt.figure(figsize=(10,7))
sns.scatterplot(x="total_bill",y="tip",hue="day",style="time",size="size",data=tips)
plt.xlabel("Total Bill")
plt.show()
```
